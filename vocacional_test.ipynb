```python
from dotenv import load_dotenv
load_dotenv()
```

```python
# ===============================
# üîê CONFIGURACI√ìN SEGURA
# ===============================
import os
import requests
import json
from getpass import getpass

# Cargar API Key y Endpoint si no est√°n definidos
if "AZURE_OPENAI_KEY" not in os.environ:
    os.environ["AZURE_OPENAI_KEY"] = getpass("Introduce tu API Key de Azure OpenAI: ")

if "AZURE_OPENAI_ENDPOINT" not in os.environ:
    os.environ["AZURE_OPENAI_ENDPOINT"] = input("Introduce tu Endpoint (ej. https://endpoint.services.ai.azure.com/): ")

if "AZURE_OPENAI_DEPLOYMENT" not in os.environ:
    os.environ["AZURE_OPENAI_DEPLOYMENT"] = input("Introduce el nombre del deployment (ej. openai): ")


# ===============================
# ‚öôÔ∏è CONFIGURACI√ìN BASE
# ===============================
API_KEY = os.getenv("AZURE_OPENAI_KEY")
ENDPOINT = os.getenv("AZURE_OPENAI_ENDPOINT")
DEPLOYMENT = os.getenv("AZURE_OPENAI_DEPLOYMENT")

# Endpoint final (se construye autom√°ticamente)
URL = f"{ENDPOINT}openai/deployments/{DEPLOYMENT}/chat/completions?api-version=2024-05-01-preview"


# ===============================
# üí¨ SOLICITUD A AZURE OPENAI
# ===============================
headers = {
    "Content-Type": "application/json",
    "api-key": API_KEY
}

data = {
    "messages": [
        {"role": "system", "content": "Eres un asistente educativo amable y claro."},
        {"role": "user", "content": "Explica qu√© es el aprendizaje profundo con un ejemplo sencillo."}
    ],
    "temperature": 0.7,
    "max_tokens": 300
}

response = requests.post(URL, headers=headers, data=json.dumps(data))

# ===============================
# üì§ RESULTADO
# ===============================
if response.status_code == 200:
    print(response.json()["choices"][0]["message"]["content"])
else:
    print("‚ùå Error:", response.status_code)
    print(response.text)
```
